```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('buyrentke (1).csv')
```

```python
df.head(10)
```

```python
# Split the 'Amenities' column into separate columns
amenities_split = df['Amenities'].str.split(', ', expand=True)
```

```python
# Rename the split columns for better readability
amenities_split.columns = [f'Amenity_{i+1}' for i in range(amenities_split.shape[1])]
```

```python
# Concatenate the original DataFrame with the new amenities columns
df_split = pd.concat([df.drop(columns=['Amenities']), amenities_split], axis=1)
```
